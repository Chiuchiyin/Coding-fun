## Intermediate Level

### Part 3: Calculate the average price of an order.

Hint: Examine the data to see if the `quantity` column is relevant to this calculation.

Hint: Think carefully about the simplest way to do this!

```python
#find the number of unique order_ids's
orders = []
for row in data:
    if row[0] in row:
        orders.append(row[0])
num_orders = len(orders)
num_orders
```

```python
# A list of prices
prices = []
for row in data:
    if row[4] in row:
        prices.append(row[4][1:-1]) # remove dollar sign and end space 
```

```python
float_prices = []
for item in prices:
    float_prices.append(float(item))
```

```python
# average price of an order.
sum(float_prices) / num_orders
```

### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
burrito_toppings = []
for row in data:
    if 'Burrito' in row[2]:
        burrito_toppings.append(row[3])
#burrito_toppings
```

```python
burrito_count = len(burrito_toppings)
burrito_count
```

```python
# Number of toppings by counting commas amd adding 1
toppings_count = 0
for row in data:
    if 'Burrito' in row[2]:
        toppings_count += (row[3].count(',') + 1)
toppings_count
```

```python
# average number of toppings per burrito
toppings_count / burrito_count
```
